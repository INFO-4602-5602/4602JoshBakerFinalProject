```python
import pandas as pd
```

```python
educDF = pd.read_csv('Education Index.csv', encoding='latin1')
```

```python
educDF.head()
```

```python
educDF = educDF.drop(['HDI Rank (2018)'], axis=1)
educDF = educDF.melt(id_vars=['Country'], var_name='Year', value_name='Value')
```

```python
educDF.head()
```

```python
educDF = educDF.replace({'..': None})
educDF = educDF.dropna()
educDF.head()
```

```python
educDF.to_csv('New Education.csv',index=False)
```
